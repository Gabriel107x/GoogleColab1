```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y= np.sin(x)
```

```python
fig, axes = plt.subplots(1,2,figsize=(10,5))
#Plot1
axes[0].plot(x,y, label="$sin(x)$")
axes[0].set_title('Relation X - Y')
axes[0].set_xlabel('x axis')
axes[0].set_ylabel('y axis')
axes[0].legend()
#plot2
axes[1].plot(y,x)
axes[1].set_title('Relation Y - X')
axes[1].set_xlabel('y axis')
axes[1].set_ylabel('x axis')
```

```python
plt.plot(x,y,label='sin(x)')
plt.title('Este es un titulo')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend(loc='lower left', bbox_to_anchor=(1,0.9))
```
