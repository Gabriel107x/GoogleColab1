```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips.head(2)
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip');
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=tips, x='total_bill', y='tip',hue='day', style='time', size='size')
plt.legend(loc='center',bbox_to_anchor=(1.12,0.5));
```

```python
plt.figure(figsize=(9,5))
markers = {"Lunch":"D", "Dinner":"s"}
sns.scatterplot(data=tips, x='total_bill', y='tip',hue='day', style='time', size='size', markers=markers)
plt.legend(loc='center',bbox_to_anchor=(1.12,0.5));
```

```python
sns.lineplot(data=tips, x='total_bill', y='tip');
```

```python
sns.lineplot(data=tips, x='total_bill', y='tip',hue='day', style='time', size='size')
plt.legend(loc='center',bbox_to_anchor=(1.12,0.5))
;
```

```python
sns.relplot(data=tips, x='total_bill', y='tip',hue='day', style='time', size='size', col='day')
;
```

```python
fig, axes = plt.subplots(2,2, figsize=(10,10))

sns.lineplot(ax=axes[0,0], data=tips, x='total_bill', y='tip')
sns.scatterplot(ax=axes[1,0], data=tips, x='total_bill', y='tip')
sns.lineplot(ax=axes[0,1], data=tips, x='total_bill', y='tip',hue='day', style='time', size='size')
plt.legend(loc='center',bbox_to_anchor=(1.4,0.5))

markers = {"Lunch":"D", "Dinner":"s"}
sns.scatterplot(ax=axes[1,1], data=tips, x='total_bill', y='tip',hue='day', style='time', size='size', markers=markers)
plt.legend(loc='center',bbox_to_anchor=(1.2,0.5))
```
