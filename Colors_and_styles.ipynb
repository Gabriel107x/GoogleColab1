```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
```

```python
print(plt.style.available)
```

```python
plt.style.use('dark_background')
```

## Matlab

```python
fig, ax = plt.subplots(figsize=(8,8))
ax.plot(x,x+1,'r--')
ax.plot(x,x+2,'bx-')
ax.plot(x,x+3,'y.-')
ax.plot(x,x+4,'go:')
```

##Pyplot

```python
fig, ax = plt.subplots(figsize=(8,8))
ax.plot(x,x+1,color='green', alpha=0.6, linewidth=5, linestyle='-', marker='o', markersize='10', markerfacecolor='red')
ax.plot(x,x+2,color='blue', linewidth=8, linestyle='--', marker='x')
ax.plot(x,x+3,color='#66ff66', linewidth=2, linestyle='dashed', marker='v')
ax.plot(x,x+4,color='#ccffcc', linewidth=0.5, linestyle=':', marker='P')
```

```python
fig, ax = plt.subplots(figsize=(8,8))
ax.plot(x,x+1,'r')
ax.plot(x,x+2,'b')
ax.plot(x,x+3,'y')
ax.plot(x,x+4,'g')
```
